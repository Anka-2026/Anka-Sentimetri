```python
class AnkaAnalizor:
    def __init__(self):
        # Temel Sözlükler
        self.duygu_sozlugu = {"hızlı":2, "çözüm":3, "kaba":-3, "yavaş":-2, "iyi":1, "rezalet":-5}
        self.dereceler = {"çok":2, "fazla":2}
        self.olumsuzluklar = ["değil", "yok", "hiç"]
        
        # YENİ: Aciliyet Kelimeleri
        self.kritik_kelimeler = ["iade", "avukat", "şikayet", "bekliyorum", "param"]
        
        print("Anka-Sentimetri V1.5 Başlatıldı! 🚀")

    def analiz_et(self, metin):
        puan = 0
        carpan = 1
        olumsuzluk_dugmesi = False
        aciliyet_durumu = False # Yeni şalterimiz
        
        kelimeler = metin.lower().split()
        
        # 1. Adım: Aciliyet Kontrolü
        for kelime in kelimeler:
            if kelime in self.kritik_kelimeler:
                aciliyet_durumu = True
        
        # 2. Adım: Duygu Analizi (Senin o meşhur tersten döngün)
        for k in kelimeler[::-1]:
            if k in self.olumsuzluklar:
                olumsuzluk_dugmesi = True
            elif k in self.dereceler:
                carpan = self.dereceler[k]
            elif k in self.duygu_sozlugu:
                temel_puan = self.duygu_sozlugu[k]
                hesaplanan = temel_puan * carpan
                
                if olumsuzluk_dugmesi:
                    puan += (hesaplanan * -1)
                    olumsuzluk_dugmesi = False
                else:
                    puan += hesaplanan
                carpan = 1
        
        # Sonucu sözlük olarak döndürelim (Daha profesyonel!)
        return {"skor": puan, "kritik": aciliyet_durumu}

    # 3. Adım: Toplu Raporlama Yeteneği
    def rapor_hazirla(self, yorum_listesi):
        print("\n--- ANKA KALİTE DENETİM RAPORU ---")
        for yorum in yorum_listesi:
            sonuc = self.analiz_et(yorum)
            durum = "🚩 KRİTİK" if sonuc['kritik'] else "✅ Normal"
            duygu = "😊 Pozitif" if sonuc['skor'] > 0 else "☹️ Negatif"
            
            print(f"Yorum: {yorum[:30]}... | Puan: {sonuc['skor']} | {durum} | {duygu}")
```

```python
# Makineyi hazırla
servis = AnkaAnalizor()

# Test verilerimiz (Çağrı merkezinden gelmiş gibi)
gelen_yorumlar = [
    "Hızlı çözüm için teşekkürler",
    "Param iade edilmedi çok yavaş",
    "Kaba bir davranış değil ama çok yavaş",
    "Avukatımla görüştüm şikayetçiyim"
]

# Raporu çalıştır
servis.rapor_hazirla(gelen_yorumlar)
```
